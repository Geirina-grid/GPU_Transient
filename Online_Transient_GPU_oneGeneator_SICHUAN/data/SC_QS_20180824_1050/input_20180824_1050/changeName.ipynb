```python
import csv  
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
acline.to_csv('ACline.csv',mode='w',index=False)
```

```python
bus = pd.read_csv('Bus.csv')
firstJ = bus["name"].str.split(".").str[1]
secondJ = bus["name"].str.split(".").str[2]
nameJ = firstJ.str.cat(secondJ, sep='.')
bus['name'] = pd.DataFrame({'list':nameJ.values})
```

```python
bus.to_csv('Bus.csv',mode='w',index=False)
```
